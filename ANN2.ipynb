2. Write a Python program to plot a few activation functions that are being used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def linear(x):
    return x
```

```python
# Generate x values
x = np.linspace(-5, 5, 100)

# Plot each activation function on a separate graph
plt.figure(figsize=(10, 6))
```

```python
# Sigmoid
plt.subplot(2, 2,1)
plt.plot(x, sigmoid(x), color='blue')
plt.title('Sigmoid')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
```

```python
# Tanh
plt.subplot(2, 2, 2)
plt.plot(x, tanh(x), color='red')
plt.title('Tanh')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
```

```python
# ReLU
plt.subplot(2, 2, 3)
plt.plot(x, relu(x), color='green')
plt.title('ReLU')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
```

```python
# Linear
plt.subplot(2, 2, 4)
plt.plot(x, linear(x), color='purple')
plt.title('Linear')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
```

```python
plt.tight_layout()
plt.show()
```
