Write a python Program for Bidirectional Associative Memory with two pairs of vectors.

```python
import numpy as np

class BAM:
    def __init__(self, input_size, output_size):
        self.input_size = input_size
        self.output_size = output_size
        self.W = np.zeros((input_size, output_size))

    def train(self, X, Y):
        self.W += np.dot(X.T, Y)

    def recall(self, X):
        return np.dot(X, self.W)

# Define input and output pairs
X1 = np.array([[1, 1, -1]])
Y1 = np.array([[1, -1]])

X2 = np.array([[-1, -1, 1]])
Y2 = np.array([[-1, 1]])

# Initialize BAM with input and output sizes
input_size = X1.shape[1]
output_size = Y1.shape[1]
bam = BAM(input_size, output_size)

# Train the BAM with the pairs
bam.train(X1, Y1)
bam.train(X2, Y2)

# Recall the output for each input
output1 = bam.recall(X1)
output2 = bam.recall(X2)

print("Input 1:", X1, "Output 1:", output1)
print("Input 2:", X2, "Output 2:", output2)
```
